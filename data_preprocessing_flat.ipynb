```python
import pandas as pd
import re
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("/content/flats.csv")
```

```python
df.sample(5)
```

```python
df.shape
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
#check for missing values
df.isnull().sum()
```

```python
#dropping irrelevant columns from data
#columns dropped - link, property_id

df.drop(["link","property_id"],inplace=True,axis=1)
```

```python
df.head(3)
```

```python
df.rename(columns={'area':'price_per_sqft'},inplace=True)
```

```python
#Society column
df.society.value_counts().sample(5)
```

```python
#removing the rating value from the society name and converting them to lower case

df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★','',str(name)).strip()).str.lower()
```

```python
df.society.value_counts().sample(5)
```

```python
#price column
#assigning a common unit to price
#removing the rows where the value is 'Price on Request'

df[df['price'] == 'Price on Request'].shape
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
df.shape
```

```python
# function to convert price to one single unit
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df["price"].sample(5)
```

```python
df.head()
```

```python
#cleaning the price per sqft column
#extract the numeric price of each flat

df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```

```python
df.head(3)
```

```python
#bedroom bathroom and baclony colums

df.bedRoom.value_counts()
```

```python
df.bedRoom.isnull().sum(), df.bathroom.isnull().sum(), df.balcony.isnull().sum()
```

```python
#Extracting the number of bedrooms
df['bedRoom'] = df['bedRoom'].str.split(' ').str.get(0).astype('int')
df.head(2)
```

```python
#Extracting the number of bathrooms

df['bathroom'] = df['bathroom'].str.split(' ').str.get(0).astype('int')
df.head(2)
```

```python
#Extracting the balcony values
df['balcony'] = df['balcony'].str.split(' ').str.get(0).str.replace('No','0')
df.head(2)
```

```python
df.balcony.value_counts()
```

```python
# additionalRoom column
df['additionalRoom'].value_counts()
```

```python
#checking null values
df['additionalRoom'].isnull().sum()
```

```python
#replacing the null values with not available
df['additionalRoom'].fillna('not available',inplace=True)
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df.head(3)
```

```python
#floorNum column - keeping the floor where the flat is

df['floorNum'] = df['floorNum'].str.split(' ').str.get(0).replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract(r'(\d+)')
```

```python
df.head(2)
```

```python
# facing column
df['facing'].value_counts()
```

```python
df['facing'].isnull().sum()
```

```python
#replacing the null facing values with NA
df['facing'].fillna('NA',inplace=True)
```

```python
df.head()
```

```python
#Creating an area attribute using the price per sqfeet and price columns

df['area'] = round((df['price']*10000000)/df['price_per_sqft'])
```

```python
df.insert(4,'area',round((df['price']*10000000)/df['price_per_sqft']))
df.head()
```

```python
df.insert(loc=1,column='property_type',value='flat')
df.head(2)
```

```python
df.shape
```

```python
#saving the cleaned flat file

df.to_csv('flats_cleaned.csv',index=False)
```
